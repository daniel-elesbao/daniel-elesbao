```python
# Gerando a matriz do grafo que representa as rodovias federais que ligam as capitais do Brasil entre si
import pandas as pd
from google.colab import files

# Define as capitais do Brasil como vértices do grafo
capitais = ["Rio Branco", "Maceió", "Macapá", "Manaus", "Salvador", "Fortaleza", "Brasília", "Vitória", "Goiânia", "São Luís", "Cuiabá", "Campo Grande", "Belo Horizonte", "Belém", "João Pessoa", "Curitiba", "Recife", "Teresina", "Rio de Janeiro", "Natal", "Porto Alegre", "Porto Velho", "Boa Vista", "Florianópolis", "São Paulo", "Aracaju", "Palmas"]

#cria um dataframe pandas, inicialmente com zeros
caprod = pd.DataFrame({'capitais':capitais})
for item in capitais:
  caprod[item] = [0]*27

# Define as rodovias federais que ligam as capitais entre si como arestas do grafo
# Fonte: https://www.gov.br/transportes/pt-br/assuntos/transporte-terrestre/rodovias-federais/mapas
rodovias = [("Rio Branco", "Porto Velho", "BR-364"), ("Maceió", "Recife", "BR-101"), ("Maceió", "Aracaju", "BR-101"), ("Macapá", "Belém", "BR-156"), ("Manaus", "Boa Vista", "BR-174"), ("Manaus", "Porto Velho", "BR-319"), ("Salvador", "Aracaju", "BR-101"), ("Salvador", "Maceió", "BR-101"), ("Salvador", "Recife", "BR-101"), ("Salvador", "Vitória", "BR-101"), ("Salvador", "Belo Horizonte", "BR-116"), ("Fortaleza", "Teresina", "BR-020"), ("Fortaleza", "Natal", "BR-304"), ("Fortaleza", "Recife", "BR-116"), ("Brasília", "Goiânia", "BR-060"), ("Brasília", "Belo Horizonte", "BR-040"), ("Brasília", "Palmas", "BR-010"), ("Brasília", "Cuiabá", "BR-070"), ("Brasília", "Campo Grande", "BR-060"), ("Brasília", "São Paulo", "BR-050"), ("Brasília", "Salvador", "BR-020"), ("Brasília", "Fortaleza", "BR-020"), ("Vitória", "Belo Horizonte", "BR-262"), ("Vitória", "Rio de Janeiro", "BR-101"), ("Goiânia", "Cuiabá", "BR-070"), ("Goiânia", "Campo Grande", "BR-060"), ("Goiânia", "Belo Horizonte", "BR-050"), ("Goiânia", "Palmas", "BR-153"), ("São Luís", "Teresina", "BR-135"), ("São Luís", "Belém", "BR-316"), ("Cuiabá", "Campo Grande", "BR-163"), ("Cuiabá", "Porto Velho", "BR-364"), ("Campo Grande", "São Paulo", "BR-267"), ("Campo Grande", "Curitiba", "BR-163"), ("Belo Horizonte", "Rio de Janeiro", "BR-040"), ("Belo Horizonte", "São Paulo", "BR-381"), ("Belo Horizonte", "Curitiba", "BR-262"), ("Belém", "Macapá", "BR-156"), ("Belém", "Palmas", "BR-010"), ("João Pessoa", "Recife", "BR-101"), ("João Pessoa", "Natal", "BR-101"), ("Curitiba", "Florianópolis", "BR-101"), ("Curitiba", "São Paulo", "BR-116"), ("Curitiba", "Porto Alegre", "BR-116"), ("Recife", "Natal", "BR-101"), ("Rio de Janeiro", "São Paulo", "BR-116"), ("Rio de Janeiro", "Vitória", "BR-101"), ("Natal", "Fortaleza", "BR-304"), ("Porto Alegre", "Florianópolis", "BR-101"), ("Porto Velho", "Boa Vista", "BR-174"), ("Florianópolis", "São Paulo", "BR-101")]

# Cria uma matriz 27 x 27, inicialmente com zeros
matriz = []
for i in range(27):
  linha = []
  for j in range(27):
    linha.append(0)
  matriz.append(linha)

# Preenchendo a matriz com 1 onde há uma rodovia federal direta entre duas capitais
for item in rodovias:
  capital1 = item[0]
  capital2 = item[1]
  # Obtendo os índices das capitais na lista de vértices
  index1 = capitais.index(capital1)
  index2 = capitais.index(capital2)
  # Atribuindo 1 na posição correspontente do DataFrame Pandas
  caprod.loc[index2,capital1] = [1]
  caprod.loc[index1,capital2] = [1]

  # Atribuindo 1 na posição correspondente da matriz
  matriz[index1][index2] = 1
  matriz[index2][index1] = 1 # Como as rodovias ligam as capitais nos dois sentidos o grafo é não direcionado e a matriz é simétrica

#Adiciona informações de latitude e longitude das capitais
#Fonte: https://geoftp.ibge.gov.br/organizacao_do_territorio/estrutura_territorial/localidades/
caprod['latitude'] = [-9.97, -9.66, 0.04, -3.1, -12.97, -3.72, -15.78, -20.32, -16.68, -2.53, -15.6, -20.45, -19.92, -1.46, -7.12, -25.42, -8.06, -5.09, -22.91, -5.79, -30.03, -8.76, 2.82, -27.59, -23.55, -10.91, -10.25]
caprod['longitude'] = [-67.81, -35.73, -51.05, -60.02, -38.51, -38.54, -47.93, -40.34, -49.25, -44.3, -56.1, -54.64, -43.94, -48.5, -34.86, -49.27, -34.88, -42.8, -43.2, -35.21, -51.23, -63.9, -60.67, -48.55, -46.64, -37.07, -48.33]

#exibe o DataFrame
display(caprod)

#exporta o DataFrame como csv
caprod.to_csv('caprod.csv', encoding = 'utf-8-sig')
files.download('caprod.csv')

# Imprime a matriz com os nomes das capitais
print("  ", end="")
for c in capitais:
  print(c, end=" ") # Usando as três primeiras letras de cada capital
print()
for i in range(27):
  print(capitais[i][:3], end=" ")
  for j in range(27):
    print(matriz[i][j], end=" ")
  print()
```
